### BCSC

```python
import json
from sklearn.metrics import f1_score
from sklearn.preprocessing import LabelEncoder

with open('../data/results_oph/BCSC_result.json', 'r') as file:
    data = json.load(file)
```

```python
tested_classes = list(data[0].keys())[3:]
# print(len(tested_classes))

for case in tested_classes:
    correct, incorrect = 0, 0
    unknown = 0
    true_labels = []
    predicted_labels = []
    for problem in data:
        true_labels.append(problem['answer'])
        if problem[case] is None or len(problem[case]) == 0:
            predicted_labels.append('Other')
            # predicted_labels.append(problem['answer'])
        elif problem[case][0] in ['A', 'B', 'C', 'D']:
                predicted_labels.append(problem[case][0])
        else:
            # print(problem[case])
            predicted_labels.append('Other')
    grader = [true_labels[i] == predicted_labels[i] for i in range(len(predicted_labels))]
    accuracy = sum(grader) / len(grader)
    label_encoder = LabelEncoder()
    label_encoder.fit(true_labels+predicted_labels)
    true_encoded = label_encoder.transform(true_labels)
    predicted_encoded = label_encoder.transform(predicted_labels)
    macro_f1 = f1_score(true_encoded, predicted_encoded, average='macro')
    print(case + ':', accuracy, macro_f1)
```
